```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, value):
        if not self.head:
            self.head = DNode(value)
        else: 
            current = self.head
            while current.next:
                current = current.next
            new_node = DNode(value)
            current.next = new_node
            new_node.prev = current
    
    def __str__(self):
        out = "[%s" % self.head.data
        current = self.head.next
        while current.next:
            out += ", %s" % current.data
            current = current.next
        out += "]"
        return out
    
    def get_tail(self):
        current = self.head
        while current.next:
            current = current.next
        return current
    
    def __len__(self):
        counter = 0
        current = self.head
        while current:
            current = current.next
            counter += 1
        return counter
            

    # create a method that allows us to print our list backwards.
    # IE if our list is [0,1,2,3]; then our new method should return 
    # [3,2,1,0]    
    def print_inverted(self):
        current = self.get_tail()
        out = "[%s" % current.data
        current = current.prev
        while current:
            out += ", %s" % current.data
            current = current.prev
        out += "]"
        return out
    
    # Create an "insert" method which takes 2 parameters: index anf value
    # the method must insert a new node BEFORE the target index with value as its data
    #Remeber: we should be able to insert anywhere in the list
    # and that include replacing the head of our list
        
            
    def insert(self, index, value):
        if index < 0:
            raise IndexError("Negative Index")
        new_node = DNode(value)
        if index == 0:
            temp = self.head
            self.head = new_node
            new_node.next = temp
        else: 
            current = self.head
            counter = 0
            while current.next and index != counter:
                current = current.next
                counter += 1
            if index > counter:
                current.next = new_node
                new_node.prev = current
                #this means we've exhausted our list but
                # the target index is greater than the length of our list
            elif index == counter:
                prev_node = current.prev
                prev_node.next = new_node
                new_node.prev = prev_node
                new_node.next = current
                current.prev = new_node
                
        
```

```python
from functools import lru_cache

@lru_cache
def fib(n):
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)

def fib_n(n):
    if n < 2:
        return n
    return fib_n(n-1) + fib_n(n-2)
```
